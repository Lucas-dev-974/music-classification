### Retrait de tout les colonnes qui ne sont pas nécessaire pour l'entraînement

- **danceability**: La mesure de la propension de la chanson à être dansante peut être pertinente pour la classification de genre, car certains genres sont plus propices à la danse que d'autres.
- **energy**: L'énergie perçue dans la chanson peut aider à distinguer des genres avec différents niveaux d'intensité.
- **key**: La tonalité peut influencer le caractère général de la musique, ce qui pourrait être pertinent pour la classification de genre.
- **loudness**: Le volume global perçu peut être un facteur dans la distinction entre les genres.
- **mode**: Le mode musical (majeur ou mineur) peut également influencer l'atmosphère de la chanson et donc sa classification.
- **tempo**: Le tempo de la chanson est important pour de nombreux genres musicaux.
- **time_signature**: La signature rythmique peut aider à différencier les genres ayant des structures rythmiques spécifiques.

```python

selected_features = ['danceability', 'energy', 'key', 'loudness', 'mode', 'tempo', 'time_signature','track_id']
df = df[selected_features].dropna()

df['track_id'] = df['track_id'].astype(str).str.strip("b''")

df.head(50)
```

## Ajout des genres

```python
labels_dict = {}

label_file_path = './Data/msd_tagtraum_cd2c.cls'

with open(label_file_path, 'r') as cls_file:
    for i in  range(7):
        next(cls_file)

    for line in cls_file:
        parts = line.strip().split()
        track_id = parts[0]
        genres = parts[1:]  # Prendre tous les éléments après le premier comme genres

        labels_dict[track_id] = genres

data = [(track_id, genre[0]) for track_id, genre in labels_dict.items()]
df_genres = pd.DataFrame(data, columns=['track_id', 'genre'])

df_genres.head()
```

### Retrait de tout les champs qui ne sont pas nécessaire pour l'entraînement

```python
df = df.merge(df_genres, on='track_id', how='left')
df.head(50)
```

```python
null_counts = df.isnull().sum()
null_counts
```

```python
csv_file_path = 'songs.csv'
df.to_csv(csv_file_path, index=False, encoding='utf-8')
```
